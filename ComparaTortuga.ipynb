```python
#Importamos las librerias necesarias
import cv2
import numpy as np
import os
```

```python
#Cargamos las dos imagenes donde buscar y comparar el patrón

original = cv2.imread("tortuga1.JPG")
image_to_compare = cv2.imread("tortuga2.JPG")
```

```python
#  Primero verificamos si las imagenes son la misma
if original.shape == image_to_compare.shape:
    print("Las imagenes tienen el mismo tamaño y canales")
    difference = cv2.subtract(original, image_to_compare)
    b, g, r = cv2.split(difference)
    if cv2.countNonZero(b) == 0 and cv2.countNonZero(g) == 0 and cv2.countNonZero(r) == 0:
        print("Las imagenes son completamente iguales")
    else:
        print("Las imagenes tienen Shape igual pero son diferentes")
else:
        print("Las imagenes tienen Shape diferente")
```

```python
# Si no son iguales estudiamos las similitudes

sift = cv2.xfeatures2d.SIFT_create()
kp_1, desc_1 = sift.detectAndCompute(original, None)
kp_2, desc_2 = sift.detectAndCompute(image_to_compare, None)

index_params = dict(algorithm=0, trees=5)
search_params = dict()
flann = cv2.FlannBasedMatcher(index_params, search_params)

matches = flann.knnMatch(desc_1, desc_2, k=2)

good_points = []
for m, n in matches:
    if m.distance < 0.6*n.distance:
        good_points.append(m)

# Define how similar they are
number_keypoints = 0
if len(kp_1) <= len(kp_2):
    number_keypoints = len(kp_1)
else:
    number_keypoints = len(kp_2)

if len(good_points)>30:
    print("Puntuación de matches:", len(good_points),"Puntuación de similitud: ", len(good_points) / number_keypoints * 100,"MATCH!****************************" )
    result = cv2.drawMatches(original, kp_1, image_to_compare, kp_2, good_points, None)
    cv2.imwrite("feature_matching.jpg", result)
else:
    print("No Match")    


```
